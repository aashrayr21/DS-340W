```python
import warnings
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

try:
    import seaborn as sns
    sns.set_style("whitegrid")
except Exception:
    pass

import networkx as nx

import statsmodels.api as sm
from statsmodels.tsa.api import VAR
from statsmodels.regression.linear_model import OLS
from scipy.stats import skew, kurtosis, jarque_bera
from statsmodels.regression.quantile_regression import QuantReg

warnings.filterwarnings("ignore")
plt.rcParams["figure.figsize"] = (12, 5)

ROOT       = Path.cwd()                         
BASEDIR    = ROOT / "raw_yf"
START, END = "2005-01-01", "2018-12-31"

OUTDIR    = ROOT / "oil_outputs"
CONN_OUT  = ROOT / "connectedness_outputs"
TABDIR    = CONN_OUT / "tables"
for p in (OUTDIR, CONN_OUT, TABDIR):
    p.mkdir(parents=True, exist_ok=True)

print(f"ROOT = {ROOT}")
print(f"Outputs → {OUTDIR.resolve()}, {CONN_OUT.resolve()}")

VAR_LAGS           = 2       
FEVD_H             = 10      
DY_MAX_K           = 12      
ROLL_WINDOW_WISH   = 200     
PRINT_ROWS         = 40      
COVERAGE_MIN_SHARE = 0.90    
EXCLUDE_OIL_IN_TABLE2 = True 




def _unique_monotonic(df_or_s):
    obj = df_or_s.copy()
    obj = obj[~obj.index.duplicated(keep="last")]
    obj = obj.sort_index()
    return obj

def load_price_flexible(csv_path: Path, start: str = START, end: str = END):
 
    if not csv_path.exists():
        raise FileNotFoundError(f"CSV not found: {csv_path}")
    df = pd.read_csv(csv_path, parse_dates=["Date"]).sort_values("Date")
    price_cols = [c for c in ("Adj Close","AdjClose","Close","Price","PX_LAST","Value","Last","Close Price")
                  if c in df.columns]
    if not price_cols:
        price_cols = [c for c in df.columns if c != "Date" and pd.api.types.is_numeric_dtype(df[c])]
    if not price_cols:
        raise ValueError(f"{csv_path.name}: no usable price column.")
    s = df.set_index("Date")[price_cols[0]].astype(float)
    s = s.loc[start:end]
    s = s[~s.index.duplicated(keep="last")].sort_index()
    bday_idx = pd.bdate_range(start, end)
    s = s.reindex(bday_idx).ffill()
    s.name = csv_path.stem
    return s

def load_panel(csv_path: Path, start: str = START, end: str = END):

    if not csv_path.exists():
        raise FileNotFoundError(f"CSV not found: {csv_path}")
    df = pd.read_csv(csv_path, parse_dates=["Date"]).sort_values("Date").set_index("Date")
    df = df.loc[start:end]
    df = df.select_dtypes(include=["number"]).astype(float)
    df = df[~df.index.duplicated(keep="last")].sort_index()
    bday_idx = pd.bdate_range(start, end)
    df = df.reindex(bday_idx).ffill()
    return df

def log_ret(obj):
    return np.log(obj / obj.shift(1))

def pct(x): 
    return 100 * x

def latex_table(df: pd.DataFrame, path: Path, caption: str, label: str, index: bool = True):

    try:
        with open(path, "w") as f:
            f.write("\\begin{table}[!ht]\n\\centering\n")
            f.write(df.to_latex(index=index, escape=True))
            f.write(f"\n\\caption{{{caption}}}\n\\label{{{label}}}\n\\end{{table}}\n")
        print(f"  LaTeX saved → {path}")
    except Exception as e:
        print(f"⚠️  Failed to save LaTeX to {path}: {e}")




try:
    oil = load_price_flexible(BASEDIR / "BZ=F_2005_2018.csv").rename("Oil")
    vix = load_price_flexible(BASEDIR / "^VIX_2005_2018.csv").rename("VIX")

    xle_path = BASEDIR / "XLE_2005_2018.csv"
    xle = load_price_flexible(xle_path).rename("XLE") if xle_path.is_file() else None

    exp_prices = load_panel(BASEDIR / "exporter_sectors_2005_2018.csv")
    imp_prices = load_panel(BASEDIR / "importer_sectors_2005_2018.csv")

    SECTOR_TICKERS = ["XLE","XLB","XLI","XLY","XLP","XLV","XLF","XLK","XLC","XLU","XLRE"]
    sector_series = {}
    missing = []
    for tkr in SECTOR_TICKERS:
        fp = BASEDIR / f"{tkr}_2005_2018.csv"
        if fp.is_file():
            try:
                sector_series[tkr] = load_price_flexible(fp).rename(tkr)
            except Exception as e:
                print(f"[WARN] failed to load {tkr}: {e}")
        else:
            missing.append(tkr)
    if missing:
        print(f"[WARN] missing sector CSVs: {missing}")

    shocks_path = BASEDIR / "ready_shocks_2005_2018.csv"
    if shocks_path.is_file():
        shocks = (pd.read_csv(shocks_path, parse_dates=[0], index_col=0)
                    .reindex(pd.bdate_range(START, END))
                    .dropna())
    else:
        print("[WARN] ready_shocks not found → building AR(1) proxy from oil returns.")
        oil_ret_tmp = log_ret(oil).dropna()
        ar = sm.tsa.AutoReg(oil_ret_tmp, lags=1, old_names=False).fit()
        shocks = pd.DataFrame(index=oil_ret_tmp.index,
                              data={"supply":0.0, "demand":0.0, "risk":ar.resid})

    oil   = _unique_monotonic(oil)
    vix   = _unique_monotonic(vix)
    xle   = _unique_monotonic(xle) if xle is not None else None
    exp_prices = _unique_monotonic(exp_prices)
    imp_prices = _unique_monotonic(imp_prices)
    for k in list(sector_series.keys()):
        sector_series[k] = _unique_monotonic(sector_series[k])
    shocks = _unique_monotonic(shocks)

except FileNotFoundError as e:
    print(f" Error loading data: {e}")
    print(f"Please ensure data files exist in: {BASEDIR}")
    raise




ret_oil = log_ret(oil).rename("Oil_ret")
ret_vix = log_ret(vix).rename("VIX_ret")
ret_xle = log_ret(xle).rename("XLE_ret") if xle is not None else None

exp_ret = log_ret(exp_prices).add_suffix("_ret")
imp_ret = log_ret(imp_prices).add_suffix("_ret")
etf_ret = (pd.concat([log_ret(s).rename(f"{s.name}_ret") for s in sector_series.values()],
                     axis=1) if sector_series else pd.DataFrame())

pieces = [ret_oil, ret_vix, exp_ret, imp_ret, etf_ret, shocks]
if ret_xle is not None:
    pieces.insert(2, ret_xle)

all_ret = pd.concat(pieces, axis=1, sort=False)

all_ret = all_ret.loc[:, ~all_ret.columns.duplicated(keep='last')]

all_ret = all_ret.replace([np.inf, -np.inf], np.nan)
all_ret = _unique_monotonic(all_ret)  
print(f"Combined return matrix → {all_ret.shape[0]} dates × {all_ret.shape[1]} series")




def descriptive_table(df: pd.DataFrame) -> pd.DataFrame:
    out = {}
    for col in df.columns:
        s = df[col].dropna()
        if s.empty or len(s) < 2:
            continue
        try:
            jb, jp = jarque_bera(s.values)
        except Exception:
            jb, jp = np.nan, np.nan
        out[col] = {
            "Mean (%)"      : pct(s.mean()),
            "Std. dev. (%)" : pct(s.std(ddof=1)),
            "Skewness"      : skew(s, bias=False),
            "Kurtosis"      : kurtosis(s, fisher=False, bias=False),
            "Jarque-Bera"   : jb,
            "JB p-value"    : jp,
            "N"             : int(len(s))
        }
    return pd.DataFrame(out).T

tbl1 = descriptive_table(all_ret).round(4)

print("\nTable 1 • Descriptive statistics")
if len(tbl1) <= PRINT_ROWS:
    print(tbl1.to_string())
else:
    print(tbl1.head(PRINT_ROWS).to_string())
    print(f"... ({len(tbl1)-PRINT_ROWS} more rows not printed)")

tbl1.to_csv(TABDIR / "Table1_All_descriptives.csv")
latex_table(tbl1, TABDIR / "Table1_All_descriptives.tex",
            caption="Descriptive statistics for the full set of daily log-returns (oil, VIX, sectors, shocks).",
            label="tab:desc_all")




def ew_market(df: pd.DataFrame) -> pd.Series:
    return df.mean(axis=1)

def run_unified_panel(df_ret: pd.DataFrame, shocks_df: pd.DataFrame, label: str = "All") -> pd.DataFrame:

    df_ret     = _unique_monotonic(df_ret)
    shocks_df  = _unique_monotonic(shocks_df)

    drop_cols = [c for c in ["supply","demand","risk"] if c in df_ret.columns]
    if EXCLUDE_OIL_IN_TABLE2 and "Oil_ret" in df_ret.columns:
        drop_cols.append("Oil_ret")
    df_ret = df_ret.drop(columns=drop_cols, errors="ignore")

    results = []
    for col in sorted(df_ret.columns):
        y = df_ret[col].to_frame("y")
        other_cols = [c for c in df_ret.columns if c != col]
        market = df_ret[other_cols].mean(axis=1).to_frame("mkt")

        X = pd.concat([shocks_df[['supply','demand','risk']], market], axis=1)
        X = sm.add_constant(X)

        combo = pd.concat([y, X], axis=1, join="inner").dropna()
        if len(combo) < 20:
            continue

        y_aligned = combo["y"].values
        X_aligned = combo.drop(columns=["y"]).values

        try:
            res = OLS(y_aligned, X_aligned).fit()
            names = combo.drop(columns=["y"]).columns
            params = dict(zip(names, res.params))

            try:
                r2 = float(res.rsquared)
            except Exception:
                yhat = res.predict(X_aligned)
                ssr = np.sum((y_aligned - yhat)**2)
                sst = np.sum((y_aligned - y_aligned.mean())**2)
                r2 = 1 - ssr/sst if sst != 0 else np.nan

            results.append(pd.Series({
                "Constant"     : params.get("const", np.nan),
                "Supply shock" : params.get("supply", np.nan),
                "Demand shock" : params.get("demand", np.nan),
                "Risk shock"   : params.get("risk", np.nan),
                "Agg. market"  : params.get("mkt", np.nan),
                "R-squared"    : r2
            }, name=col))
        except Exception as e:
            print(f"[WARN] Regression failed for {col}: {e}")
            continue

    if not results:
        print(f"[WARN] No successful regressions for {label}")
        return pd.DataFrame()
    
    out = pd.DataFrame(results).round(4)

    print(f"\nTable 2 • Panel regressions ({label})")
    if len(out) <= PRINT_ROWS:
        print(out.to_string())
    else:
        print(out.head(PRINT_ROWS).to_string())
        print(f"... ({len(out)-PRINT_ROWS} more rows not printed)")

    out.to_csv(TABDIR / f"Table2_{label}_panel.csv")
    latex_table(out, TABDIR / f"Table2_{label}_panel.tex",
                caption=f"Oil-price shocks + common market factor regressions for all series ({label}).",
                label=f"tab:{label.lower()}_panel")
    print(f"  Table 2 saved for {label} ({len(out)} equations).")
    return out

tbl2 = run_unified_panel(
    all_ret.drop(columns=[c for c in ["supply","demand","risk"] if c in all_ret.columns], errors="ignore"),
    shocks, label="All"
)




def _var_to_ma(var_res, H: int) -> np.ndarray:

    K = var_res.neqs
    p = var_res.k_ar

    A = np.zeros((K * p, K * p))
    A[:K, :K * p] = np.hstack([var_res.coefs[i].T for i in range(p)])
    if p > 1:
        A[K:, :-K] = np.eye(K * (p - 1))

    C_top = np.zeros((K, K * p))
    C_top[:, :K] = np.eye(K)

    Phi = np.zeros((H, K, K))
    A_pow = np.eye(K * p)
    for h in range(H):
        if h == 0:
            Phi[h] = np.eye(K)
        else:
            A_pow = A_pow @ A
            Phi[h] = C_top @ A_pow @ C_top.T
    return Phi

def _gfevd_pesaran_shin(Phi: np.ndarray, Sigma: np.ndarray) -> np.ndarray:

    H, K, _ = Phi.shape
    Sigma = np.asarray(Sigma, dtype=float)
    theta = np.zeros((K, K))

    for i in range(K):
        den = 0.0
        for h in range(H):
            vi = Phi[h][i, :] @ Sigma
            den += vi @ Phi[h][i, :].T
        if den == 0:
            continue
        for j in range(K):
            if Sigma[j, j] == 0:
                continue
            num = 0.0
            for h in range(H):
                num += (Phi[h][i, :] @ Sigma[:, j]) ** 2
            theta[i, j] = num / Sigma[j, j]
        theta[i, :] = theta[i, :] / den

    theta = np.clip(theta, 0.0, 1.0)
    rowsum = theta.sum(axis=1, keepdims=True)
    rowsum[rowsum == 0] = 1.0
    theta = theta / rowsum
    return theta



def build_dy_panel(all_df: pd.DataFrame, max_k: int = DY_MAX_K,
                   base_keep: tuple = ("Oil_ret","VIX_ret","XLE_ret")) -> pd.DataFrame:

    df = all_df.drop(columns=[c for c in ["supply","demand","risk"] if c in all_df.columns],
                     errors="ignore")
    df = df.apply(pd.to_numeric, errors="coerce").replace([np.inf, -np.inf], np.nan)

    cover = df.notna().mean(axis=0)
    df = df.loc[:, cover >= COVERAGE_MIN_SHARE]

    df = _unique_monotonic(df).dropna(how="any")

    keep = [c for c in base_keep if c in df.columns]
    remaining = [c for c in df.columns if c not in keep]
    var_ranked = df[remaining].var().sort_values(ascending=False).index.tolist()
    for c in var_ranked:
        if len(keep) >= max_k:
            break
        keep.append(c)
    panel = df[keep].copy()
    print(f"[DY panel] Using {panel.shape[1]} series, {panel.shape[0]} rows.")
    print(f"[DY panel] Columns: {list(panel.columns)}")
    return panel

dy_df = build_dy_panel(all_ret, max_k=DY_MAX_K)

def dy_fevd_and_tci(df: pd.DataFrame, lags: int = VAR_LAGS, H: int = FEVD_H) -> dict:
    clean = df.dropna().astype(float)
    if len(clean) < lags + 50:
        raise ValueError(f"Insufficient data: need at least {lags + 50} rows, got {len(clean)}")
    var_res = VAR(clean).fit(lags)
    Sigma = np.asarray(var_res.sigma_u, dtype=float)
    Phi   = _var_to_ma(var_res, H)
    theta = _gfevd_pesaran_shin(Phi, Sigma)

    fevd = pd.DataFrame(theta * 100, index=clean.columns, columns=clean.columns)
    theta_off = theta.copy()
    np.fill_diagonal(theta_off, 0.0)
    TCI = theta_off.sum().sum() / theta.shape[0] * 100.0

    TO   = fevd.sum(axis=1) - np.diag(fevd)
    FROM = fevd.sum(axis=0) - np.diag(fevd)
    NET  = TO - FROM
    return {"FEVD": fevd, "TCI": TCI, "NET": NET, "VAR": var_res}

try:
    dy_full = dy_fevd_and_tci(dy_df, lags=VAR_LAGS, H=FEVD_H)
    print(f"\nFull-sample Total Connectedness Index (TCI) = {dy_full['TCI']:.2f}%")
    dy_full["FEVD"].to_csv(CONN_OUT / "FEVD_full_system_compact.csv")
except Exception as e:
    print(f" Error computing full-sample DY: {e}")
    dy_full = None




def rolling_tci_with_diag(df: pd.DataFrame, window: int, lags: int = VAR_LAGS, H: int = FEVD_H) -> pd.Series:
    vals, dates, fails = [], [], 0
    for i in range(window, len(df) + 1):
        sub = df.iloc[i - window:i]
        try:
            tci = dy_fevd_and_tci(sub, lags=lags, H=H)['TCI']
            if not np.isfinite(tci):
                raise ValueError("Non-finite TCI")
        except Exception as e:
            tci = np.nan
            fails += 1
        vals.append(tci)
        dates.append(df.index[i - 1])
    s = pd.Series(vals, index=pd.DatetimeIndex(dates), name="TCI")
    ok = s.notna().sum()
    print(f"[Rolling TCI] windows={len(s)}, ok={ok}, failed={fails}, ok%={100*ok/max(len(s),1):.1f}%")
    return s

if dy_full is not None:
    N = len(dy_df)
    EFFECTIVE_ROLL_WINDOW = min(ROLL_WINDOW_WISH, max(50, int(0.6 * N)))
    print(f"[Rolling TCI] Using window={EFFECTIVE_ROLL_WINDOW} (N={N})")

    tci_series = rolling_tci_with_diag(dy_df, window=EFFECTIVE_ROLL_WINDOW, lags=VAR_LAGS, H=FEVD_H)
    tci_series.to_csv(CONN_OUT / "Fig1_TCI_compact_series.csv")

    plt.figure(figsize=(12,4))
    plt.plot(tci_series.index, tci_series.values, lw=1.2)
    plt.title(f"Rolling Total Connectedness Index ({EFFECTIVE_ROLL_WINDOW}-day window) – Compact System")
    plt.ylabel("TCI (%)"); plt.xlabel("Date"); plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.savefig(CONN_OUT / "Fig1_timevarying_connectedness_compact.png", dpi=300, bbox_inches="tight")
    plt.show()
    print("  Figure 1 saved.")



def band_across_params(df: pd.DataFrame, lags_list=None, H_list=None, window: int = None):
    assert (lags_list is not None) ^ (H_list is not None), "Supply either lags_list or H_list."
    if window is None:
        N = len(df)
        window = min(ROLL_WINDOW_WISH, max(50, int(0.6 * N)))
    grid = {}
    if lags_list:
        for p in lags_list:
            try:
                grid[f"L{p}"] = rolling_tci_with_diag(df, window=window, lags=p, H=FEVD_H)
            except Exception as e:
                print(f"[WARN] Failed for lag={p}: {e}")
    else:
        for h in H_list:
            try:
                grid[f"H{h}"] = rolling_tci_with_diag(df, window=window, lags=VAR_LAGS, H=h)
            except Exception as e:
                print(f"[WARN] Failed for H={h}: {e}")
    if not grid:
        return pd.DataFrame(), pd.DataFrame()
    grid_df = pd.concat(grid, axis=1)
    if grid_df.dropna(how="all").empty:
        return pd.DataFrame(), grid_df
    band = pd.DataFrame({
        "lower" : grid_df.quantile(0.1, axis=1),
        "median": grid_df.quantile(0.5, axis=1),
        "upper" : grid_df.quantile(0.9, axis=1)
    })
    return band, grid_df

def plot_band(band: pd.DataFrame, title: str, ylims=(84,92), fname: Path | None = None):
    if band.empty:
        print("[WARN] band is empty → nothing to plot.")
        return
    plt.figure(figsize=(12,4))
    plt.fill_between(band.index, band["lower"], band["upper"], alpha=0.2, label="10–90% band")
    plt.plot(band.index, band["median"], lw=1.2, label="Median")
    plt.ylim(*ylims); plt.title(title); plt.ylabel("TCI (%)"); plt.grid(alpha=0.3); plt.legend()
    plt.tight_layout()
    if fname:
        plt.savefig(fname, dpi=300, bbox_inches="tight")
        print(f"  Band plot saved → {fname}")
    plt.show()

if dy_full is not None:
    band_lag, _ = band_across_params(dy_df, lags_list=[1,2,3,4])
    if not band_lag.empty:
        band_lag.to_csv(CONN_OUT / "Sensitivity_lags_1to4_compact.csv")
        plot_band(band_lag, "Lag-order sensitivity (L=1–4) – Compact System",
                  fname=CONN_OUT / "Appendix_LagBand_Compact.png")

    band_h, _ = band_across_params(dy_df, H_list=list(range(5,16)))
    if not band_h.empty:
        band_h.to_csv(CONN_OUT / "Sensitivity_H_5to15_compact.csv")
        plot_band(band_h, "Forecast-horizon sensitivity (H=5–15) – Compact System",
                  fname=CONN_OUT / "Appendix_HBand_Compact.png")




def plot_network(fevd: pd.DataFrame,
                 threshold: float = 0.03,
                 title: str = "Connectedness Network (≥3% spillovers)",
                 fname: Path | None = None):
    G = nx.DiGraph()
    colour_map = {}
    for node in fevd.columns:
        if node.startswith(("supply","demand","risk")):
            colour_map[node] = "#ff7f0e"
        elif any(k in node for k in ("Oil","VIX","XLE")):
            colour_map[node] = "#2ca02c"
        elif node.startswith("X"):
            colour_map[node] = "#1f77b4"
        elif node.startswith(("E","I")):
            colour_map[node] = "#d62728"
        else:
            colour_map[node] = "#8c564b"
        G.add_node(node, colour=colour_map[node])

    for i, src in enumerate(fevd.index):
        for j, dst in enumerate(fevd.columns):
            if i == j:
                continue
            w = fevd.iloc[i, j] / 100.0
            if w >= threshold:
                G.add_edge(src, dst, weight=w)

    if G.number_of_nodes() == 0:
        print("[WARN] Network has no nodes")
        return
        
    pos = nx.spring_layout(G, seed=42, k=0.5)
    edge_weights = [5 * G[u][v]["weight"] for u, v in G.edges()]
    node_colours = [data["colour"] for _, data in G.nodes(data=True)]

    plt.figure(figsize=(12,9))
    nx.draw_networkx_nodes(G, pos, node_size=900, node_color=node_colours, alpha=0.9)
    nx.draw_networkx_labels(G, pos, font_size=9)
    nx.draw_networkx_edges(G, pos, arrowstyle='->', arrowsize=12, width=edge_weights,
                           edge_color="#555555", alpha=0.7)
    plt.title(title); plt.axis("off"); plt.tight_layout()
    if fname:
        plt.savefig(fname, dpi=300, bbox_inches="tight")
        print(f" Network graph saved → {fname}")
    plt.show()

if dy_full is not None:
    plot_network(dy_full["FEVD"],
                 threshold=0.03,
                 title="Full-system Diebold–Yilmaz Network (≥3% spillovers)",
                 fname=CONN_OUT / "Network_FullSystem_Compact.png")




def _drop_zero_var_cols(X: pd.DataFrame) -> pd.DataFrame:
    var = X.var(axis=0)
    keep = var.index[var > 1e-12]
    return X[keep]

def quantile_var_proxy(df: pd.DataFrame, tau: float = 0.05, lags: int = VAR_LAGS, H: int = FEVD_H) -> dict:

    df = df.dropna().astype(float)
    K = df.shape[1]
    if K < 2:
        raise ValueError("Need at least 2 series for QVAR.")

    Y = df.iloc[lags:].copy()
    X_lags = [df.shift(h) for h in range(1, lags + 1)]
    X = pd.concat(X_lags, axis=1).iloc[lags:].copy()
    X.columns = [f"{col}_L{h}" for h in range(1, lags + 1) for col in df.columns]
    X.insert(0, "const", 1.0)

    common_idx = Y.index.intersection(X.index)
    Y = Y.loc[common_idx]
    X = X.loc[common_idx]
    X = _drop_zero_var_cols(X)
    mask = X.notna().all(axis=1) & Y.notna().all(axis=1)
    X, Y = X[mask], Y[mask]

    if len(Y) < 50:
        raise ValueError(f"[QVAR τ={tau:.2f}] Too few rows ({len(Y)}).")

    Xv = X.values
    fitted = np.empty_like(Y.values, dtype=float)
    coefs = {}

    for k, col in enumerate(Y.columns):
        yk = Y[col].values.reshape(-1)  
        try:
            qr = QuantReg(yk, Xv)
            res = qr.fit(q=tau, max_iter=2000)
            fitted[:, k] = res.predict(Xv).reshape(-1)
            coefs[col] = res.params
        except Exception as e:
            ols_res = OLS(yk, Xv).fit()
            fitted[:, k] = ols_res.predict(Xv).reshape(-1)
            coefs[col] = ols_res.params

    fitted_df = pd.DataFrame(fitted, index=Y.index, columns=Y.columns)

    var_res = VAR(fitted_df).fit(lags)
    Sigma = np.asarray(var_res.sigma_u, dtype=float)
    Phi   = _var_to_ma(var_res, H)
    theta = _gfevd_pesaran_shin(Phi, Sigma)

    fevd = pd.DataFrame(theta * 100, index=Y.columns, columns=Y.columns)
    theta_off = theta.copy()
    np.fill_diagonal(theta_off, 0.0)
    TCI = theta_off.sum().sum() / theta.shape[0] * 100.0
    TO   = fevd.sum(axis=1) - np.diag(fevd)
    FROM = fevd.sum(axis=0) - np.diag(fevd)
    NET  = TO - FROM

    return {"FEVD": fevd, "TCI": TCI, "NET": NET, "coef": coefs, "tau": tau, "VAR": var_res}

if dy_full is not None:
    try:
        q05 = quantile_var_proxy(dy_df, tau=0.05, lags=VAR_LAGS, H=FEVD_H)
        print(f"\n[QVAR] τ=0.05  → TCI = {q05['TCI']:.2f}%")
        q05["FEVD"].to_csv(CONN_OUT / "FEVD_QVAR_tau005.csv")

        q95 = quantile_var_proxy(dy_df, tau=0.95, lags=VAR_LAGS, H=FEVD_H)
        print(f"[QVAR] τ=0.95  → TCI = {q95['TCI']:.2f}%")
        q95["FEVD"].to_csv(CONN_OUT / "FEVD_QVAR_tau095.csv")

        print("\nQVAR FEVD (τ=0.05) — head")
        print(q05["FEVD"].round(2).head(PRINT_ROWS).to_string())
        print("\nQVAR FEVD (τ=0.95) — head")
        print(q95["FEVD"].round(2).head(PRINT_ROWS).to_string())

        latex_table(q05["FEVD"].round(2), TABDIR / "FEVD_QVAR_tau005.tex",
                    caption="Quantile-VAR proxy FEVD at τ=0.05 (left tail).",
                    label="tab:fevd_qvar_tau005")
        latex_table(q95["FEVD"].round(2), TABDIR / "FEVD_QVAR_tau095.tex",
                    caption="Quantile-VAR proxy FEVD at τ=0.95 (right tail).",
                    label="tab:fevd_qvar_tau095")

        try:
            q50 = quantile_var_proxy(dy_df, tau=0.50, lags=VAR_LAGS, H=FEVD_H)
            print(f"[QVAR] τ=0.50  → TCI = {q50['TCI']:.2f}%")
            pd.Series({"tau=0.05": q05["TCI"], "tau=0.50": q50["TCI"], "tau=0.95": q95["TCI"]})\
              .round(2).to_csv(CONN_OUT / "TCI_QVAR_tails_vs_median.csv")
        except Exception as e:
            print(f"[QVAR] median τ=0.50 run skipped: {e}")

        plot_network(q05["FEVD"], threshold=0.03,
                     title="QVAR τ=0.05 — Network (≥3% spillovers)",
                     fname=CONN_OUT / "Network_QVAR_tau005.png")
        plot_network(q95["FEVD"], threshold=0.03,
                     title="QVAR τ=0.95 — Network (≥3% spillovers)",
                     fname=CONN_OUT / "Network_QVAR_tau095.png")
    except Exception as e:
        print(f" QVAR analysis failed: {e}")


def detect_structural_breaks_icss(series, max_breaks=5):
    series_clean = series.dropna()
    n = len(series_clean)
    if n < 60:
        return []
    values = series_clean.values
    css = np.cumsum(values**2)
    normalized_css = css / css[-1]
    k = np.arange(1, n+1)
    D = np.abs(normalized_css - k/n)
    D_max = np.max(D)
    critical_value = 1.358 / np.sqrt(n)  

    breaks = []
    if D_max > critical_value:
        segments = [(0, n-1)]
        for _ in range(max_breaks):
            changed = False
            new_segments = []
            for start, end in segments:
                if end - start < 20:
                    new_segments.append((start, end))
                    continue
                seg = values[start:end+1]
                seg_css = np.cumsum(seg**2)
                seg_norm = seg_css / seg_css[-1]
                sk = np.arange(1, len(seg)+1)
                Dseg = np.abs(seg_norm - sk/len(seg))
                bidx = int(np.argmax(Dseg))
                if Dseg[bidx] > critical_value:
                    actual = start + bidx
                    breaks.append(series_clean.index[actual])
                    new_segments.extend([(start, actual), (actual+1, end)])
                    changed = True
                else:
                    new_segments.append((start, end))
            segments = new_segments
            if not changed:
                break
    return sorted(set(breaks))


def bai_perron_multiple_breaks(y, x=None, max_breaks=5, min_segment=25, fcrit=6.0):

    y = y.dropna()
    if len(y) < 3*min_segment:
        return []

    if x is None:
        X = pd.DataFrame({"const": 1.0}, index=y.index)
    else:
        X = (x if isinstance(x, pd.DataFrame) else pd.DataFrame(x, index=y.index)).copy()
        if "const" not in X.columns:
            X.insert(0, "const", 1.0)

    idx = y.index.intersection(X.index)
    y, X = y.loc[idx], X.loc[idx]

    breaks = []

    def _fit(yseg, xseg):
        try:
            return OLS(yseg.values, xseg.values).fit()
        except Exception:
            return None

    def _search(start, end):
        if end - start + 1 < 2*min_segment or len(breaks) >= max_breaks:
            return
        yseg = y.iloc[start:end+1]
        xseg = X.iloc[start:end+1]
        pool = _fit(yseg, xseg)
        if pool is None:
            return
        ssr_pool = pool.ssr
        k = xseg.shape[1]
        nseg = len(yseg)

        best_f, best_c = -np.inf, None
        for c in range(start + min_segment, end - min_segment + 1):
            y1, x1 = y.iloc[start:c], X.iloc[start:c]
            y2, x2 = y.iloc[c:end+1], X.iloc[c:end+1]
            r1, r2 = _fit(y1, x1), _fit(y2, x2)
            if r1 is None or r2 is None:
                continue
            ssr1, ssr2 = r1.ssr, r2.ssr
            denom_df = max(nseg - 2*k, 1)
            fval = ((ssr_pool - (ssr1 + ssr2)) / k) / ((ssr1 + ssr2) / denom_df)
            if fval > best_f:
                best_f, best_c = fval, c

        if best_c is not None and best_f > fcrit:
            breaks.append(y.index[best_c])
            _search(start, best_c - 1)
            _search(best_c, end)

    _search(0, len(y) - 1)
    return sorted(set(breaks))


def plot_structural_breaks(series, breaks, title="Structural Breaks"):
    plt.figure(figsize=(12, 4))
    plt.plot(series.index, series.values, label='Series')
    for i, brk in enumerate(breaks):
        plt.axvline(x=brk, color='red', linestyle='--', alpha=0.7,
                    label='Break' if i == 0 else "")
    plt.title(f"{title} (Detected: {len(breaks)} breaks)")
    plt.legend()
    plt.tight_layout()
    return plt.gcf()


print("\n" + "="*60)
print("STRUCTURAL BREAKS ANALYSIS")
print("="*60)

key_series = {
    'VIX_ret': all_ret['VIX_ret'],
    'Oil_ret': all_ret['Oil_ret']
}
if 'tci_series' in locals() and tci_series is not None:
    key_series['TCI_series'] = tci_series

structural_breaks_results = {}
for name, series in key_series.items():
    if series is None or series.dropna().empty:
        continue
    print(f"\n{name} Structural Breaks")
    try:
        variance_breaks = detect_structural_breaks_icss(series)
        print(f"Variance breaks (ICSS): {len(variance_breaks)}")
        for brk in variance_breaks:
            print(f"  {brk.strftime('%Y-%m-%d')}")
    except Exception as e:
        print(f"Variance breaks failed: {e}")
        variance_breaks = []

    try:
        mean_breaks = bai_perron_multiple_breaks(series)
        print(f"Mean breaks (Bai-Perron): {len(mean_breaks)}")
        for brk in mean_breaks:
            print(f"  {brk.strftime('%Y-%m-%d')}")
    except Exception as e:
        print(f"Mean breaks failed: {e}")
        mean_breaks = []

    all_b = sorted(set(variance_breaks + mean_breaks))
    structural_breaks_results[name] = {
        'variance_breaks': variance_breaks,
        'mean_breaks': mean_breaks,
        'all_breaks': all_b
    }
    if all_b:
        try:
            fig = plot_structural_breaks(series, all_b, f"{name} - Structural Breaks")
            plt.savefig(CONN_OUT / f"StructuralBreaks_{name}.png", dpi=300, bbox_inches='tight')
            plt.close()
            print(f"  Plot saved: StructuralBreaks_{name}.png")
        except Exception as e:
            print(f"  Plot failed: {e}")

rows = []
for name, res in structural_breaks_results.items():
    for kind in ['variance_breaks', 'mean_breaks']:
        for brk in res[kind]:
            rows.append({'series': name, 'break_type': kind.replace('_breaks', ''), 'date': brk})
if rows:
    pd.DataFrame(rows).to_csv(CONN_OUT / "StructuralBreaks_Summary.csv", index=False)
    print(f"\nStructural breaks summary saved with {len(rows)} breaks")
else:
    print("\nNo structural breaks detected")


def rolling_tvp_var_analysis_fixed(df, window=252, step=22):
    dates, tci_values = [], []
    avail = df.columns.tolist()
    desired = ['VIX_ret', 'Oil_ret', 'XLE_ret', 'XLF_ret']
    cols = [c for c in desired if c in avail]
    if len(cols) < 2:
        print(f"Insufficient common columns. Available: {avail}")
        return None
    tvp_df = df[cols].dropna()

    for i in range(window, len(tvp_df), step):
        sub = tvp_df.iloc[i-window:i]
        try:
            res = dy_fevd_and_tci(sub, lags=min(VAR_LAGS, 2), H=min(FEVD_H, 8))
            tci_values.append(res['TCI'])
            dates.append(tvp_df.index[i-1])
        except Exception:
            continue

    if not tci_values:
        return None
    return pd.Series(tci_values, index=dates, name='Rolling_TVP_TCI')


def summarize_static_vs_dynamic(dy_full, rolling_tvp):
    static_tci = float(dy_full['TCI'])
    dyn_mean = float(rolling_tvp.mean())
    dyn_min, dyn_max = float(rolling_tvp.min()), float(rolling_tvp.max())
    gap = static_tci - dyn_mean
    gap_pct = 100 * gap / static_tci if static_tci else np.nan
    print(f"\nStatic TCI: {static_tci:.2f}% | Dynamic mean: {dyn_mean:.2f}% "
          f"(range {dyn_min:.2f}–{dyn_max:.2f}%). Gap: {gap:.2f}% ({gap_pct:.1f}%).")
    return {"static": static_tci, "dyn_mean": dyn_mean, "dyn_min": dyn_min,
            "dyn_max": dyn_max, "gap": gap, "gap_pct": gap_pct}


print("\n" + "="*60)
print("TIME-VARYING PARAMETER VAR ANALYSIS (Fixed)")
print("="*60)

try:
    rolling_tvp = rolling_tvp_var_analysis_fixed(dy_df, window=200, step=22)
    if rolling_tvp is not None and not rolling_tvp.dropna().empty:
        plt.figure(figsize=(12, 4))
        plt.plot(rolling_tvp.index, rolling_tvp.values, color='darkred', linewidth=1.5)
        plt.title('Time-Varying Total Connectedness Index (Rolling TVP-VAR)')
        plt.ylabel('TCI (%)')
        plt.grid(alpha=0.3)
        plt.tight_layout()
        plt.savefig(CONN_OUT / "TVP_VAR_TCI_Fixed.png", dpi=300, bbox_inches='tight')
        plt.show()
        rolling_tvp.to_csv(CONN_OUT / "TVP_VAR_TCI_Fixed.csv")
        _tvps = summarize_static_vs_dynamic(dy_full, rolling_tvp)
    else:
        print("Rolling TVP-VAR failed - insufficient successful windows")
except Exception as e:
    print(f"TVP-VAR analysis failed: {e}")


def _stationary_bootstrap(df, block_size):
    n = len(df)
    idx = []
    i = 0
    while i < n:
        L = np.random.geometric(1/block_size) + 1
        start = np.random.randint(0, n)
        for j in range(L):
            if i >= n:
                break
            idx.append((start + j) % n)
            i += 1
    return df.iloc[idx].reset_index(drop=True)


def bootstrap_confidence_intervals(df, func, n_bootstrap=500, ci=0.95, **kwargs):
    try:
        original = func(df, **kwargs)
    except Exception as e:
        print(f"Original function failed: {e}")
        return np.nan, (np.nan, np.nan)

    vals = []
    for _ in range(n_bootstrap):
        try:
            bs = _stationary_bootstrap(df, block_size=max(5, int(len(df)**0.33)))
            if len(bs) < 100:
                continue
            val = func(bs, **kwargs)
            if np.isfinite(val):
                vals.append(val)
        except Exception:
            continue

    if len(vals) < 50:
        return original, (np.nan, np.nan)

    vals = np.array(vals)
    lo = np.percentile(vals, (1-ci)*50)
    hi = np.percentile(vals, (1+ci)*50)
    return original, (lo, hi)


def test_connectedness_differences(df1, df2, n_bootstrap=500):
    def tci(df):
        return dy_fevd_and_tci(df, lags=VAR_LAGS, H=FEVD_H)['TCI']

    diff0 = tci(df1) - tci(df2)
    if not np.isfinite(diff0):
        return {'difference': np.nan, 'p_value': np.nan, 'significant': False}

    pooled = pd.concat([df1, df2])
    diffs = []
    for _ in range(n_bootstrap):
        try:
            s1 = _stationary_bootstrap(pooled, 10).iloc[:len(df1)]
            s2 = _stationary_bootstrap(pooled, 10).iloc[:len(df2)]
            d = tci(s1) - tci(s2)
            if np.isfinite(d):
                diffs.append(d)
        except Exception:
            continue

    if len(diffs) < 100:
        return {'difference': diff0, 'p_value': np.nan, 'significant': False}

    diffs = np.array(diffs)
    pval = np.mean(np.abs(diffs) >= np.abs(diff0))
    return {'difference': diff0, 'p_value': pval, 'significant': pval < 0.05}


print("\n" + "="*60)
print("ECONOMIC SIGNIFICANCE TESTS")
print("="*60)

print("\nBootstrap Confidence Intervals")
try:
    tci_orig = dy_full['TCI']
    tci_stat, (lo, hi) = bootstrap_confidence_intervals(
        dy_df,
        lambda x: dy_fevd_and_tci(x, lags=VAR_LAGS, H=FEVD_H)['TCI'],
        n_bootstrap=500
    )
    print(f"TCI: {tci_orig:.2f}%")
    if np.isfinite(lo) and np.isfinite(hi):
        print(f"95% CI: [{lo:.2f}%, {hi:.2f}%]")
    else:
        print("95% CI: Could not compute")
except Exception as e:
    print(f"Bootstrap CI failed: {e}")

if structural_breaks_results.get('VIX_ret', {}).get('all_breaks'):
    main_break = structural_breaks_results['VIX_ret']['all_breaks'][0]
    pre = dy_df[dy_df.index < main_break]
    post = dy_df[dy_df.index >= main_break]
    if len(pre) > 100 and len(post) > 100:
        print(f"\nRegime Change Test (Break: {main_break.strftime('%Y-%m-%d')})")
        rt = test_connectedness_differences(pre, post, n_bootstrap=200)
        print(f"TCI Difference: {rt['difference']:.2f}%")
        print(f"P-value: {rt['p_value']:.4f}")
        print(f"Significant: {rt['significant']}")


def analyze_frequency_effects(df, frequencies=('D', 'W', 'M')):
    results = {}
    for freq in frequencies:
        print(f"\n{freq} Frequency Analysis")
        try:
            resampled = df if freq == 'D' else df.resample(freq).last().dropna()
            if len(resampled) < 100:
                print(f"  Insufficient data after {freq} resampling: {len(resampled)} observations")
                continue
            lags = 1 if freq != 'D' else VAR_LAGS
            H = 5 if freq != 'D' else FEVD_H
            conn = dy_fevd_and_tci(resampled, lags=lags, H=H)
            results[freq] = {'TCI': conn['TCI'], 'n_obs': len(resampled), 'FEVD': conn['FEVD']}
            print(f"  TCI: {conn['TCI']:.2f}%")
            print(f"  Observations: {len(resampled)}")
        except Exception as e:
            print(f"  Failed for {freq}: {e}")
    return results


def realized_volatility_connectedness(df, window=22):
    try:
        rv = df.rolling(window=window, min_periods=int(window*0.8)).std().replace(0, np.nan).dropna()
        if len(rv) < 100:
            print("Insufficient data for volatility connectedness")
            return None
        vol_conn = dy_fevd_and_tci(rv, lags=1, H=5)
        ret_conn = dy_fevd_and_tci(df.reindex(rv.index).dropna(), lags=VAR_LAGS, H=FEVD_H)
        return {
            'volatility_TCI': vol_conn['TCI'],
            'return_TCI': ret_conn['TCI'],
            'difference': vol_conn['TCI'] - ret_conn['TCI'],
            'volatility_FEVD': vol_conn['FEVD'],
            'return_FEVD': ret_conn['FEVD'],
        }
    except Exception as e:
        print(f"Volatility connectedness failed: {e}")
        return None


print("\n" + "="*60)
print("FREQUENCY EFFECTS ANALYSIS")
print("="*60)

try:
    frequency_results = analyze_frequency_effects(dy_df, frequencies=('D', 'W', 'M'))
except Exception as e:
    print(f"Frequency analysis failed: {e}")
    frequency_results = {}

print("\nVolatility vs Return Connectedness")
try:
    vol_results = realized_volatility_connectedness(dy_df)
    if vol_results:
        print(f"Volatility TCI: {vol_results['volatility_TCI']:.2f}%")
        print(f"Return TCI: {vol_results['return_TCI']:.2f}%")
        print(f"Difference: {vol_results['difference']:.2f}%")
    else:
        print("Volatility connectedness not computed")
except Exception as e:
    print(f"Volatility analysis failed: {e}")


def create_simulated_macro_factors(dates, seed=42):
    rng = np.random.default_rng(seed)
    n = len(dates)
    t = np.linspace(0, 2*np.pi, n)
    macro = pd.DataFrame({
        'term_spread': 2.0 + 0.5*np.sin(t) + 0.3*rng.normal(0, 1, n),
        'credit_spread': 1.5 + 0.3*np.cos(t) + 0.2*rng.normal(0, 1, n),
        'vix_level': 20 + 5*np.sin(2*t) + 2*rng.normal(0, 1, n),
        'usd_index': 95 + 3*np.cos(t) + 1.5*rng.normal(0, 1, n),
    }, index=dates)
    return macro


def macro_spillover_shares(fevd, fin_cols, mac_cols):
    F = fevd.copy() / 100.0
    np.fill_diagonal(F.values, 0.0)
    total_off = F.values.sum()
    if total_off == 0:
        return dict(mac_to_fin=np.nan, fin_to_macro=np.nan, fin_to_fin=np.nan, mac_to_mac=np.nan)
    mac_to_fin = F.loc[fin_cols, mac_cols].values.sum() / total_off
    fin_to_macro = F.loc[mac_cols, fin_cols].values.sum() / total_off
    fin_to_fin = F.loc[fin_cols, fin_cols].values.sum() / total_off
    mac_to_mac = F.loc[mac_cols, mac_cols].values.sum() / total_off
    return {k: 100*v for k, v in {
        "mac_to_fin": mac_to_fin,
        "fin_to_macro": fin_to_macro,
        "fin_to_fin": fin_to_fin,
        "mac_to_mac": mac_to_mac
    }.items()}


def macro_augmented_connectedness(financial_df, macro_df, lags=2, H=10):
    try:
        common = financial_df.index.intersection(macro_df.index)
        combined = pd.concat([financial_df.reindex(common), macro_df.reindex(common)], axis=1).dropna()
        if len(combined) < 100:
            print("Insufficient data for macro-augmented analysis")
            return None
        conn = dy_fevd_and_tci(combined, lags=lags, H=H)
        fin_cols = [c for c in financial_df.columns if c in combined.columns]
        mac_cols = [c for c in macro_df.columns if c in combined.columns]
        shares = macro_spillover_shares(conn['FEVD'], fin_cols, mac_cols)
        return {
            'full_TCI': conn['TCI'],
            'shares': shares,
            'FEVD': conn['FEVD'],
            'NET': conn['NET'],
            'fin_cols': fin_cols,
            'mac_cols': mac_cols
        }
    except Exception as e:
        print(f"Macro-augmented connectedness failed: {e}")
        return None


print("\n" + "="*60)
print("MACROECONOMIC INTEGRATION ANALYSIS")
print("="*60)

try:
    macro_factors = create_simulated_macro_factors(dy_df.index)
    macro_conn = macro_augmented_connectedness(dy_df, macro_factors)
    if macro_conn:
        s = macro_conn['shares']
        print(f"Full System TCI (with macro): {macro_conn['full_TCI']:.2f}%")
        print(f"Off-diag shares (%): Macro→Fin {s['mac_to_fin']:.1f}, "
              f"Fin→Macro {s['fin_to_macro']:.1f}, Fin→Fin {s['fin_to_fin']:.1f}, "
              f"Macro→Macro {s['mac_to_mac']:.1f}")

        labs = ['Macro→Financial', 'Financial→Macro', 'Financial→Financial', 'Macro→Macro']
        vals = [s['mac_to_fin'], s['fin_to_macro'], s['fin_to_fin'], s['mac_to_mac']]
        plt.figure(figsize=(8, 4))
        plt.bar(labs, vals)
        plt.title('Off-diagonal Spillover Shares (Normalized)')
        plt.ylabel('Share (%)')
        plt.tight_layout()
        plt.savefig(CONN_OUT / "Macro_Spillover_Shares.png", dpi=300, bbox_inches='tight')
        plt.show()
    else:
        print("Macro-augmented connectedness not computed")
except Exception as e:
    print(f"Macro integration failed: {e}")
```
