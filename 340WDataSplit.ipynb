```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
from sklearn.preprocessing import MinMaxScaler

CrudeOilPrices = '/content/drive/MyDrive/CrudeOilPrices1946-Present.csv'

CrudeOilPrices = pd.read_csv(CrudeOilPrices)

print(CrudeOilPrices.head())

Vix = '/content/drive/MyDrive/Vix Data 1990- Present.csv'

Vix = pd.read_csv(Vix)

print(Vix.head())
```

```python
CrudeOilPrices = CrudeOilPrices.dropna(subset=['Value'])

value_series = CrudeOilPrices['Value'].reset_index(drop=True)

df = pd.DataFrame({'Value': value_series})
df['Shift1'] = df['Value'].shift(1)
df['Shift2'] = df['Value'].shift(2)
df['Shift3'] = df['Value'].shift(3)

df['Target'] = (df['Value'].shift(-1) > df['Value']).astype(int)

df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)

total_size = len(df)
train_end = int(0.7 * total_size)
val_end = int(0.9 * total_size)

train_df = df.iloc[:train_end]
valid_df = df.iloc[train_end:val_end]
test_df  = df.iloc[val_end:]

X_train = train_df[['Shift1', 'Shift2', 'Shift3']]
y_train = train_df['Target']

X_valid = valid_df[['Shift1', 'Shift2', 'Shift3']]
y_valid = valid_df['Target']

X_test  = test_df[['Shift1', 'Shift2', 'Shift3']]
y_test  = test_df['Target']

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_valid_scaled = scaler.transform(X_valid)
X_test_scaled  = scaler.transform(X_test)

X_train_2d = X_train_scaled.copy()
X_valid_2d = X_valid_scaled.copy()
X_test_2d  = X_test_scaled.copy()

X_train_3d = X_train_scaled.reshape((X_train_scaled.shape[0], 1, X_train_scaled.shape[1]))
X_valid_3d = X_valid_scaled.reshape((X_valid_scaled.shape[0], 1, X_valid_scaled.shape[1]))
X_test_3d  = X_test_scaled.reshape((X_test_scaled.shape[0], 1, X_test_scaled.shape[1]))
```
